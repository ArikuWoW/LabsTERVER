```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Генерация данных
np.random.seed(42)
n = 100

# Доля людей 70+ (процентное соотношение)
X = np.random.uniform(10, 40, n)

# Приверженность ЗОЖ с линейной зависимостью от X и случайным шумом
Y = 0.5 * X + np.random.normal(5, 2, n)

# Рассчет корреляции и линейной регрессии
slope, intercept, r_value, p_value, std_err = linregress(X, Y)
line_fit = slope * X + intercept
r_squared = r_value**2

# Построение графиков
plt.figure(figsize=(12, 6))

# График рассеяния с линией регрессии
plt.subplot(1, 2, 1)
plt.scatter(X, Y, alpha=0.7, label="Данные")
plt.plot(X, line_fit, color='red', label=f'Линейная регрессия\n$R^2={r_squared:.2f}$')
plt.xlabel("Доля людей 70+ (%)")
plt.ylabel("Приверженность ЗОЖ")
plt.title("Гистограмма рассеяния с линией регрессии")
plt.legend()
plt.grid()

# Гистограмма распределения данных
plt.subplot(1, 2, 2)
plt.hist(X, bins=15, alpha=0.7, label="Доля людей 70+")
plt.hist(Y, bins=15, alpha=0.7, label="Приверженность ЗОЖ")
plt.xlabel("Значение")
plt.ylabel("Частота")
plt.title("Гистограммы распределений")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

r_value, r_squared
```

Данные сгенерированы и обработаны:

Коэффициент корреляции (r): 0.9183 — высокая положительная связь.
Коэффициент детерминации (R²): 0.8434 — 84.34% вариации в приверженности ЗОЖ объясняется долей людей 70+.


Далее перейдем ко второй задаче: генерации датасета, выборок и расчета доверительных интервалов.


```python
from scipy.stats import t

# Генерация датасета
regions = [f"Region_{i}" for i in range(1, 101)]
federal_districts = np.random.choice(["District_A", "District_B", "District_C"], size=n)
health_index = np.random.uniform(50, 100, n)  # Индекс ЗОЖ
elderly_percentage = np.random.uniform(10, 40, n)  # Процент людей 70+

# Создаем DataFrame
data = pd.DataFrame({
    "Region": regions,
    "Federal_District": federal_districts,
    "Health_Index": health_index,
    "Elderly_Percentage": elderly_percentage
})

# Генеральная средняя по Health_Index
population_mean = data["Health_Index"].mean()

# Случайная выборка (~1/3)
random_sample = data.sample(frac=1/3, random_state=42)
random_mean = random_sample["Health_Index"].mean()

# Стратифицированная выборка (~1/3 из каждой страты)
stratified_sample = data.groupby("Federal_District", group_keys=False).apply(
    lambda x: x.sample(frac=1/3, random_state=42)
)
stratified_mean = stratified_sample["Health_Index"].mean()

# Доверительные интервалы для среднего
confidence_levels = [0.90, 0.95, 0.99]
confidence_intervals = {
    "Random": [],
    "Stratified": []
}

def compute_confidence_interval(sample_mean, sample_std, n, confidence):
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * (sample_std / np.sqrt(n))
    return sample_mean - margin_of_error, sample_mean + margin_of_error

for confidence in confidence_levels:
    # Для случайной выборки
    random_std = random_sample["Health_Index"].std(ddof=1)
    random_n = len(random_sample)
    ci_random = compute_confidence_interval(random_mean, random_std, random_n, confidence)
    confidence_intervals["Random"].append(ci_random)
    
    # Для стратифицированной выборки
    stratified_std = stratified_sample["Health_Index"].std(ddof=1)
    stratified_n = len(stratified_sample)
    ci_stratified = compute_confidence_interval(stratified_mean, stratified_std, stratified_n, confidence)
    confidence_intervals["Stratified"].append(ci_stratified)

population_mean, random_mean, stratified_mean, confidence_intervals
```
